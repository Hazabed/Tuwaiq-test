### Identify and analyze patterns in the data.

```python
city_counts = df['City'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(city_counts.values, labels=city_counts.index, autopct='%1.1f%%',colors=sns.color_palette('pastel'))
plt.title('Distribution of Cities')
plt.show()
```

```python
city_counts = df['City'].value_counts()
plt.figure(figsize=(8, 6))
plt.bar(city_counts.index, city_counts.values)
plt.xticks(rotation=45)
plt.show()
```

```python
for col in ['TrafficIndexLive','JamsCount','JamsDelay']:
  by_city = df.groupby('City').agg(
      mean = (col,'mean'),
      median = (col,'median'),
      std = (col,'std'),
      min = (col,'min'),
      max = (col,'max')
  )
  print(col)
  print(by_city)
  print('\n')
```

### Explore relationships between variables (e.g., Traffic_Index vs. Weather_Condition).


```python
df_corr = df.copy()
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_corr['City'] = le.fit_transform(df_corr['City'])
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df_corr.corr(), annot=True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TrafficIndexLive', y='JamsDelay', data=df)
plt.title('Traffic Index vs. Jams Delay')
plt.xlabel('Traffic Index Live')
plt.ylabel('Jams Delay (minutes)')
plt.show()
```
